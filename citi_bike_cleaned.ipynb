```python
import pandas as pd
```

```python
citi_bike_feb_19 = pd.read_csv('/Users/rinabitas/Documents/Tableau-challenge/Resources/201902-citibike-tripdata.csv')
citi_bike_feb_20 = pd.read_csv('/Users/rinabitas/Documents/Tableau-challenge/Resources/202002-citibike-tripdata.csv')
```

```python
citi_bike_feb_19.head()
```

```python
citi_bike_feb_20.head()
```

```python
combined_bike_df = pd.concat([citi_bike_feb_19, citi_bike_feb_20])
combined_bike_df.head()
```

```python
combined_bike_df["gender"].replace({1: 'Male', 2: 'Female', 0: 'Unknown'}, inplace=True)
combined_bike_df.head()
```

```python
combined_bike_df['Trip Year'] = combined_bike_df.starttime.str.split("-", expand=True)[0]
combined_bike_df.head()
```

```python
clean_citi_bike = combined_bike_df
clean_citi_bike.to_csv('/Users/rinabitas/Documents/Tableau-challenge/Resources/clean_citi_bike.csv')
```
